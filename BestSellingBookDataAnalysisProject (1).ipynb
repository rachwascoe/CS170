# Did on my own (T-Test between Sales in English)

```python
#English Sales
pd.set_option('display.max_rows', None)
English_sales = data[data['Original language'] == 'English']['Approximate sales in millions']
English_sales
```

```python
#English Sales
pd.set_option('display.max_rows', None)
Russian_sales = data[data['Original language'] == 'Russian']['Approximate sales in millions']
Russian_sales
```

```python
from scipy import stats
import numpy as np

t_statistic, p_value = stats.ttest_ind(English_sales, Russian_sales, equal_var=True) 
# equal_var=True assumes equal variances; set to False for Welch's t-test

print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")

# Interpret the p-value
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the group means.")
```

# Did on my own (Grouping Genres)

```python
glist = []
for index, row in data.iterrows():
    genre=row['Genre']
    if isinstance(genre, float):
        glist.append('No Genre')
    else:
        genre=genre.lower()
        if 'fiction' in genre:
            glist.append('Fiction')
        elif 'fantasy' in genre:
           glist.append('Fantasy')
        elif 'children' in genre:
           glist.append('Children')
        elif 'novel' in genre:
           glist.append('Novel')
        else:
           glist.append('Misc')

data['Genre Groups'] = glist
data
```

```python
genre_sum = data.groupby('Genre Groups')['Approximate sales in millions'].sum().sort_values(ascending=False)
genre_sum.plot(kind='bar', color=('red', 'orange', 'yellow', 'green', 'blue', 'purple'))
plt.ylabel('Sum of approximate sales (mil)')
```
